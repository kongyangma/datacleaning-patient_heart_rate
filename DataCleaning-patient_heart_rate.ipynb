```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names= ['id', 'name', 'age', 'weight','m0006','m0612','m1218','f0006','f0612','f1218']
rate=pd.read_csv('C:/Users/kongy/Documents/Python Scripts/Datasets/patient_heart_rate.csv',names=column_names)
```

```python
rate
```

```python
#Delete Missing Values
rate.dropna(how='all',axis=0,inplace=True)
rate
```

#Problems
1: reset index, no header
2: kgs and lbs, not same measurements
3: missing values

```python
#Expand Name
rate[['First Name','Last Name']]=rate['name'].str.split(' ',expand=True)
```

```python
#Drop Column 'Name'
rate.drop(columns=['name'],axis=1, inplace=True)
```

```python
#lbs and kgs
lbs=rate['weight'].str.contains('lbs').fillna(False)
rate[lbs]
```

```python
for i,lbs_row in rate[lbs].iterrows():
    weight = int(float(lbs_row['weight'][:-3])/2.2)
    rate[lbs].at[i,'weight']='{}kgs'.format(weight)
```

```python
rate
```

```python
#Find out if there is repeated rows
rate[rate.duplicated(['age','weight','First Name'])]
```

```python
#Drop repeated rows
rate.drop_duplicates(['First Name','Last Name'],inplace=True)
rate
```

```python
# Delete Non-ASCII characters
rate['First Name'].replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
rate['Last Name'].replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
rate
```

```python
sorted_columns=['id','age','weight','First Name','Last Name']
rate=pd.melt(rate,id_vars=sorted_columns,var_name='Sex_hour',value_name='Plus_rate')
rate
```
